```python
%pip install pandas seaborn matplotlib -q
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/starbucks.csv")
df.columns = df.columns.str.strip()
df
```

```python
df.describe()
```

## Task 1: Caloric Content Across Beverage Categories

```python
df.groupby('Beverage_category')['Calories'].mean().plot(kind='barh', color='teal')
plt.figure(figsize=(10,6))
plt.title('Average Number of Calories in Each Beverage Category')
plt.xlabel('Beverage Category')
plt.ylabel('Calories')
plt.grid(True)
plt.show()
```

## Task 2: Fat Content in Different Beverages

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Total Fat (g)', y='Saturated Fat (g)', data=df, color='purple')
plt.title('Total Fat vs Saturated Fat')
plt.xlabel('Total Fat (g)')
plt.ylabel('Saturated Fat (g)')
plt.show()
```

## Task 3: Caffeine Content Variation

```python
plt.figure(figsize=(10,6))
sorted_df = df.sort_values(by='Caffeine (mg)', ascending=True)
sorted_df['Caffeine (mg)'] = pd.to_numeric(sorted_df['Caffeine (mg)'], errors='coerce')
sns.histplot(y='Beverage', x='Caffeine (mg)', data=sorted_df, color='red')
plt.title('Distribution of Caffeine Content')
plt.xlabel('Caffeine Content (mg)')
plt.ylabel('Frequency')
plt.show()
```

## Task 4: Nutritional Content Comparison

```python
plt.figure(figsize=(8,10))
sns.boxplot(x='Beverage_category', y='Protein (g)', data=df, color='green')
plt.title('Protein Per Beverage Category')
plt.xlabel('Beverage Category')
plt.ylabel('Protein (g)')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,10))
sns.boxplot(x='Beverage_category', y='Sugars (g)', data=df, color='pink')
plt.title('Sugar Per Beverage Category')
plt.xlabel('Beverage Category')
plt.ylabel('Sugar (g)')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,10))
sns.boxplot(x='Beverage_category', y='Dietary Fibre (g)', data=df, color='orange')
plt.title('Dietary Fibre Per Beverage Category')
plt.xlabel('Beverage Category')
plt.ylabel('Dietary Fibre (g)')
plt.xticks(rotation=90)
plt.show()
```

## Task 5: Vitamin and Mineral Contribution

```python
df = pd.read_csv('../data/starbucks.csv')
ax, fig = plt.subplots()
ax.barh(df['Beverage_prep'], df['Vitamin A (% DV)'], label='Vitamin A')
ax.barh(df['Beverage_prep'], df['Vitamin C (% DV)'], label='Vitamin C')
ax.barh(df['Beverage_prep'], df['Calcium (% DV)'], label='Calcium')
ax.barh(df['Beverage_prep'], df['Iron (% DV)'], label='Iron')

ax.set_xlabel('Beverage Preparation')
ax.set_title('Average Vitamin and Mineral Contribution by Beverage Preparation')
ax.legend()
plt.show()
```
