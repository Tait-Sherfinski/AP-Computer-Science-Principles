```python
%pip install pandas seaborn matplotlib -q
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/starbucks.csv")
df.columns = df.columns.str.strip()
df
```

```python
df.describe()
```

## Task 1: Caloric Content Across Beverage Categories

```python
df.groupby('Beverage_category')['Calories'].mean().plot(kind='barh', color='teal')
plt.figure(figsize=(10,6))
plt.title('Average Number of Calories in Each Beverage Category')
plt.xlabel('Beverage Category')
plt.ylabel('Calories')
plt.grid(True)
plt.show()
```

## Task 2: Fat Content in Different Beverages

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Total Fat (g)', y='Saturated Fat (g)', data=df)
plt.title('Total Fat vs Saturated Fat')
plt.xlabel('Total Fat (g)')
plt.ylabel('Saturated Fat (g)')
plt.show()
```

## Task 3: Caffeine Content Variation

```python
plt.figure(figsize=(10,6))
sns.histplot(y='Beverage', x='Caffeine (mg)', data=df.head(2500))
plt.title('Geographic Salary Distribution')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()
```
